```python
%matplotlib inline
import pandas
import numpy
import numpy.random as nrand
from PIL import Image, ImageDraw, ImageFont
from IPython.display import Image as display
import matplotlib.pyplot as plt

########################## PART ONE ######################################


WIDTH = 512 
HEIGHT = 288

# create my image
im = Image.new('RGBA', (WIDTH, HEIGHT), 'black')

# get a drawing context
draw = ImageDraw.Draw(im)

# create a number of random lines 
loop = nrand.randint(50,500)

for x in range(0, loop):
    
    #get my random points
    start_x = nrand.randint(1,WIDTH-1)
    start_y = nrand.randint(1,HEIGHT-1)
    end_x = nrand.randint(1,WIDTH-1)
    end_y = nrand.randint(1,HEIGHT-1)
    
    #get my color
    color = (nrand.randint(0,255), nrand.randint(0,255), nrand.randint(0,255), nrand.randint(0,255))
    
    #get my thickness
    line_thickness = nrand.randint(5,60)
    
    draw.line([(start_x, start_y), (end_x, end_y)], fill=color, width = line_thickness)
    
del draw

#save 
im.save('test.png')



########################## PART TWO ######################################

data = pandas.read_csv("store_train.csv")

store = data[(data.Store == 1)]

sorted_data = store.sort_values(by="Date")
#print sorted_data[["Date", "Sales"]]

dates = sorted_data["Date"].tolist()
sales = sorted_data["Sales"].tolist()

# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
# print "Current size:", fig_size
 
# Set figure width to 12 and height to 4
fig_size[0] = 12
fig_size[1] = 4
plt.rcParams["figure.figsize"] = fig_size

plt.plot_date(dates, sales, 'b-', xdate=True)

plt.xlabel("Date")
plt.ylabel("Sales $")
plt.legend(['Store #1 Sales'], loc='upper right')
plt.show()

# display saved image from part 1, plot covers it for some reason

display(filename='test.png') 

```
