```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from numpy.linalg import inv as INV

T = np.transpose
DOT = np.dot
EXP = np.exp

def Linear_Kernel(x, x_test, noise=False):
    if noise:
        if x == x_test:
            return x * x_test + 0.1
    return x * x_test

def SE_Kernel(x, x_test, noise=False):
    if noise:
        if x == x_test:
            return EXP(-0.5*(x - x_test)**2) + 0.1
    return EXP(-0.5*(x - x_test)**2)

def Poly_Kernel(x, x_test, noise=False):
    if noise:
        if x == x_test:
            return (1 + x * x_test)**3 + 0.1
    return (1 + x * x_test)**3


data_xvals = np.atleast_2d( [ 1.0, 3.0, 5.0, 6.0, 7.0, 8.0 ] )
#data_xvals = np.reshape(np.array( [ 1.0, 3.0, 5.0, 6.0, 7.0, 8.0 ] ), [6, 1])
data_yvals = np.sin( data_xvals )

# Range -2 to 10
#x_predict = [float("%.1f" % (n,)) for n in np.arange(-2, 10.1 , 0.1)]
x_predict = [n for n in np.arange(-2, 10.1 , 0.1)]
```

```python
# K(test, x) 121 x 6
def K_Test_X(x_predict, data_xvals, kernel, noise=False):
    k = [] 
    N = data_xvals.shape[1]
    for row in xrange(len(x_predict)):
        for col in xrange(N):
            k.append(kernel(x_predict[row],data_xvals[0,col], noise))
    K = np.reshape(k, (121, 6))
    return K

# K(x, x) 6 x 6
def K_X_X(data_xvals, kernel, noise=False):
    N = data_xvals.shape[1]
    K = np.reshape([ kernel(i,j,noise) for i in data_xvals[0] for j in data_xvals[0]],(N,N))
    return K

 #K(test, test) 121 x 121
def K_Test_Test(x_predict, kernel, noise=False):
    k = []
    for row in xrange(len(x_predict)):
        for col in xrange(len(x_predict)):
            k.append(kernel(x_predict[row],x_predict[col],noise))

    K = np.reshape(k, (121, 121))
    return K

# K(x, test) 6 x 121
def K_X_Test(x_predict, data_xvals, kernel, noise=False):
    k = [] # K(test, x)
    N = data_xvals.shape[1]
    for row in xrange(N):
        for col in xrange(len(x_predict)):
            k.append(kernel(x_predict[col],data_xvals[0,row], noise))

    K = np.reshape(k, (6, 121))
    return K

def Plot(x_predict, mean, sigma, data_xvals, data_yvals, title):
    a = np.reshape(x_predict, (1,121))
    b = np.reshape(mean,(1,121))
    c = np.reshape(sigma.diagonal(), (1, 121))

    plt.figure()
    plt.scatter(data_xvals[0],data_yvals[0], s=10)
    plt.plot(a[0], b[0],label='mean')
    
    a = np.reshape(x_predict, (121,1))
    b = np.reshape(mean,(121,1))
    c = np.reshape(sigma.diagonal(), (121, 1))
    
    plt.fill(np.concatenate([a, a[::-1]]),
         np.concatenate([b - c,
                        (b + c)[::-1]]),
         alpha=.3, fc='b', ec='None', label='variance')
    
    plt.grid(True, color = "gray")
    plt.title(title)
    plt.legend(loc='upper left',frameon=False)
    plt.show()
    
def Calculate(kernel, noise=False, title="My Life"):
    K_test_x = K_Test_X(x_predict, data_xvals, kernel, noise)
    K_x_x = K_X_X(data_xvals, kernel, noise)
    K_test_test = K_Test_Test(x_predict, kernel, noise)
    K_x_test = K_X_Test(x_predict, data_xvals, kernel, noise)
    
   
    
    
    K_test_test = np.reshape(np.diagonal(K_test_test), [-1, 1])
    inverse = np.linalg.pinv(K_x_x)
    
    mean = DOT( DOT( K_test_x, np.linalg.pinv(K_x_x)), T(data_yvals) )
    sigma = K_test_test - DOT( DOT( K_test_x, np.linalg.pinv(K_x_x)), K_x_test)
    print sigma.shape
    
    Plot(x_predict, mean, sigma, data_xvals, data_yvals, title)
```

```python
Calculate(Linear_Kernel, title="Linear Kernel : Noise Free")
```

```python
Calculate(Linear_Kernel, noise=True, title="Linear Kernel : Noisy")
```

```python
Calculate(Poly_Kernel, title="Polynomial Kernel : Noise Free")
```

```python
Calculate(Poly_Kernel, noise=True, title="Polynomial Kernel : Noisy")
```

```python
Calculate(SE_Kernel, title="RBF Kernel : Noise Free")
```

```python
Calculate(SE_Kernel, noise=True, title="RBF Kernel : Noisy")
```
