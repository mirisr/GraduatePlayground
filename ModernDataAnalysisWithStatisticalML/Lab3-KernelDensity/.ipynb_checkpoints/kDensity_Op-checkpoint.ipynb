```python
%matplotlib inline
import scipy.io
from matplotlib.pyplot import *
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import math
from PIL import Image
from IPython.display import Image as display
import scipy.misc
from numpy.linalg import inv
from numpy.linalg import det
from sklearn.metrics import confusion_matrix
import itertools
import time

train_mat = scipy.io.loadmat('mnist_train.mat')
train_data = train_mat['images']
train_labels = train_mat['labels']

test_mat = scipy.io.loadmat('mnist_test.mat')
test_data = test_mat['t10k_images']
test_labels = test_mat['t10k_labels']

# CONSTANTS / FUNCTIONS
# SIGMA = np.identity(784) # covariance
# INV_SIGMA = inv(SIGMA)
# DET_SIGMA = det(SIGMA)
# PI = np.pi
DOT = np.dot
TRANSPOSE = np.transpose
E = np.exp
# SQRT = math.sqrt
```

```python

def DataByLabel(x = train_data, y = train_labels):
    dataByLabels = []
    dataCounts = np.zeros(10)
    for i in xrange(10):
        condition = y == i
        indexesOfLabel, values =  np.where(condition==True)
        dataOfLabel = train_data[:,indexesOfLabel]
        dataByLabels.append(dataOfLabel)
        dataCounts[i] = dataOfLabel.shape[1]
    return dataByLabels,dataCounts

def AverageDataByLabel(dataByLabel):
    avgData = []
    for label in dataByLabel:
        labelCount = label.shape[1]
        dataSum =  np.sum(label,1)
        avgData.append( dataSum / labelCount)
    return avgData

def ShowAvgData(avgData):
    x = -1
    l = -1
    fig, ax = subplots(3, 3)
    for X in avgData:
        l+=1
        y = l % 3
        if y == 0:
            x+=1
        ax[x,y].imshow( X.reshape(28,28).T, interpolation='nearest', cmap=cm.gray)
        ax[x,y].set_axis_off()
    
    ax[2,2].set_axis_off()
    ax[2,3].set_axis_off()
    plt.show()

        
# GET REQUESITES    
dataByLabel,dataCounts = DataByLabel()

# PART 1
avgData = AverageDataByLabel(dataByLabel)
ShowAvgData(avgData)
```

```python


def PlotConfusionMatrix(cm, classes,
                          title='Confusion matrix',
                          cmap='jet'):#'PuRd'

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

def MultiGaus(x, mu, diff):
    e = E( -0.5 * DOT( TRANSPOSE(diff), (diff)) )
    #p = 0.5 **( 2 * PI * DET_SIGMA)
    return e 
```

```python




def SimplePredict(data, label, means, priors, gaus, classes=10):
    
    probs = np.zeros(classes)
    data_ = data.reshape((784,1))
    
    for class_ in xrange(classes):
        mu = means[class_]
        mu_ = mu.reshape((784, 1))
        probs[class_] += gaus(data_, mu_, (data_ - mu_)) 
    
    #probs[class_] *= priors[class_]

    predictedLabel = np.argmax(probs)
    return predictedLabel

def SimpleClassifier(means, counts):
    classes = 10
    instances = 10000
    
    # Get Priors
    priors = counts / instances
    
    # Array Holding Number of Correct predictions
    correct_pred = np.zeros(classes)
    
    # Confusion matrix init
    confusion = np.zeros((classes, classes))
    class_names = [i for i in xrange(10)]
    
    # Constanct Functions
    predict = SimplePredict
    gaus = MultiGaus
    
    # For Each Test Instance
    for instance in xrange(instances):
        label = test_labels[instance,0]
        data = test_data[:,instance]
        
        # Predict Label
        predictedLabel = predict(data,label, means, priors, gaus)
        
        # Update Prediction data
        if predictedLabel == label:
            correct_pred[label] +=1

        confusion[label, predictedLabel] += 1

        if instance % 1000 == 0:
            total = instance + 1
            print "instance:", instance, "Error Rate:", 1 - np.sum(correct_pred)/total
           
    # Plot Confusion Matrix
    print "DONE: Error Rate:", 1 - np.sum(correct_pred)/10000
    plt.figure()
    PlotConfusionMatrix(confusion, classes=class_names,
                  title='Confusion matrix Part 2')
    plt.show()




# PART 2
SimpleClassifier(avgData, dataCounts)

```
